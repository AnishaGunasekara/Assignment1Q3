```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/spider.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2RGB)
#f = cv.imread(r"E:\5th Sem\IMG Processing\assignment_01_images\spider.png", cv.IMREAD_COLOR)
#assert f is not None

x = cv.cvtColor(im_og, cv.COLOR_BGR2HSV)[:, :, 1]  # Extracting the Saturation channel
a = 0.5
sigma = 70
Y = np.minimum(x + a * 128 * np.exp(-(x - 128) ** 2 / (2 * sigma ** 2)), 255).astype('uint8')

# Creating the transformed image in HSV color space
transformed_image = cv.merge((cv.cvtColor(im_og, cv.COLOR_BGR2HSV)[:, :, 0], Y, cv.cvtColor(im_og, cv.COLOR_BGR2HSV)[:, :, 2]))

fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(cv.cvtColor(im_og, cv.COLOR_BGR2RGB))
ax[0].set_title('Original')
ax[0].axis('off')
ax[1].imshow(cv.cvtColor(transformed_image, cv.COLOR_HSV2RGB))
ax[1].set_title('Transformed (Saturation)')
ax[1].axis('off')
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

im_og = cv.imread('/content/drive/MyDrive/Image_Processing/spider.png')
im_og = cv.cvtColor(im_og, cv.COLOR_BGR2RGB)


h_im_og, s_im_og, v_im_og = cv.split(cv.cvtColor(im_og, cv.COLOR_BGR2HSV))


a = 0
sigma = 70
Y = np.minimum(s_im_og + a * 128 * np.exp(-(s_im_og - 128) ** 2 / (2 * sigma ** 2)), 255).astype('uint8')
transformed_image = cv.merge((h_im_og, Y, v_im_og))
transformed_image_bgr = cv.cvtColor(transformed_image, cv.COLOR_HSV2BGR)

vibrance_factor = 1.5
s_vibrance = np.clip(s_im_og * vibrance_factor, 0, 255).astype('uint8')
vibrance_enhanced_image = cv.merge((h_im_og, s_vibrance, v_im_og))

fig, ax= plt.subplots(1,3, figsize=(10,20))
ax[0].imshow(im_og)
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(vibrance_enhanced_image)
ax[1].set_title('Vibrance-enhanced image')
ax[1].axis('off')
ax[2].imshow(transformed_image_bgr)
ax[2].set_title('Intensity transformation')
ax[2].axis('off')
plt.show()



```
